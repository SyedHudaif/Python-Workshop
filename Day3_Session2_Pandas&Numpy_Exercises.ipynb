#Write a Pandas program to add, subtract, multiple and divide two Pandas Series and print their results using the Sample Series: [2, 4, 6, 8, ...100], [1, 3, 5, 7, 9...99]

```
import pandas as pd
ds1 = pd.Series(x for x in range(1,100) if x%2 == 0) 
ds2 = pd.Series(x for x in range(1,100) if x%2 != 0) 
print("Addition\n",ds1+ds2)
print("Subtraction\n",ds1-ds2)
print("Multiplication\n", ds1*ds2)
print("Division\n",ds1/ds2)
```

#Write a Pandas program to compare (Equivalence, greater than, less than, not equal to )the elements of the two Pandas Series.
#Sample Series: [All positive numbers divisable by 7 upto 1000], [All positive numbers divisable by 11 upto 1000]

```
import pandas as pd
ds1 = pd.Series(x for x in range(1,637) if x%7 ==0)
ds2 = pd.Series(x for x in range(1,1000) if x%11 ==0)
print("Series1:")
print(ds1)
print("Series2:")
print(ds2)
print("Compare the elements of the said Series:")
print("Equals:")
print(ds1 == ds2)
print("Greater than:")
print(ds1 > ds2)
print("Less than:")
print(ds1 < ds2)
print("Notequal:")
print(ds1 != ds2)
```

#Write a Pandas program to create a subsets of a given series based on value and condition.
#Sample NumPy array: d1 = [10, 20, 30, 40, 50... 1000]
# value and condtion :  Subset 1 :numbers divisable by 7  Subset 2 : numbers divisable by 5


```
import pandas as pd
ds = pd.Series([x for x in range(10,1000) if x%10 == 0]) 
sub1 = [x for x in ds if x%7 ==0]
print("divisible by 7\n",sub1)
sub1 = [x for x in ds if x%5 ==0]
print("divisible by 5\n",sub1)
```

#Write a Pandas program to create the mean and standard deviation, maximum and minimum of the data of a given Series of the prime numbers upto 1000

```
import pandas as pd
import sympy 
ds = pd.Series([x for x in range(0,1000) if sympy.isprime(x)])
print("Mean\n",ds.mean())
print("SD\n",ds.std())
print("Maximum value : ",ds.max())
print("Minimum value : ",ds.min())
```

#Write a Pandas program to get the elements of an array values into column-wise.
#Sample data: {X:[All odd numbers but not divisable by 7 upto 1000], 'Y':[ All even numbers but not divisable by 5 upto 1000 ],'Z':[ All odd numbers but not divisable by 3 upto 1000]}


```
import pandas as pd
ls1 = [x for x in range(0,1000) if x%2==1 and x%7!=0]
ls2 = [x for x in range(0,1000) if x%2==0 and x%5!=0]
ls3 = [x for x in range(0,1000) if x%2==1 and x%3!=0]
data = [ls1,ls2,ls3]
df = pd.DataFrame(data) 
df_t = df.T
df_t.columns = ['X', 'Y', 'Z'] 
print(df_t)
```
